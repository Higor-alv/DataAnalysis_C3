```python
# Verificar a presenÃ§a de dados ausentes
missing_data = df_train.isnull().sum()
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)
plt.figure(figsize=(10, 6))
missing_data.plot(kind='bar', color='skyblue')
plt.title('Quantidade de Dados Ausentes por VariÃ¡vel')
plt.show()
```

```python
# Verificar a distribuiÃ§Ã£o das variÃ¡veis numÃ©ricas
plt.figure(figsize=(12, 10))
sns.pairplot(df_train.select_dtypes(include=[np.number]))
plt.show()
```
